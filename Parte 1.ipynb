> b) Ahora nos gustarı́a manipular los datos, para que hagamos la predicción de la temperatura mínima para el tiempo siguiente usando la temperatura mínima de los últimos perı́odos de tiempo. El número de perı́odos de tiempos que usaremos se denomina *lag*. Por ejemplo, tendremos un *lag* igual a 3, si para predecir el valor $x_{t+1}$ en el tiempo siguiente usamos la información del tiempo actual $x_t$ y la de los dos perı́odos anteriores $x_{t-1}$ y $x_{t-2}$ como variables de entrada. La función debiese generar $(X_1,X_2,X_3)$ e $Y$:


|X0|X1|X2|Y|
|---|---|---|---|
|20.7|17.9|18.8|14.6|
|17.9|18.8|14.6|15.8|
|18.8|14.6|15.8|15.8|
|14.6|15.8|15.8|10.1|


````python
def create_dataset(dataset, lag=1):
    # Turn dataset into a row vector for easier manipulation
    dataset = dataset.reshape((len(dataset),))
    # Ammount of rows of output
    n_rows = len(dataset)-lag
    # Initialize vessels
    dataX = np.empty((n_rows, lag))
    dataY = np.empty(n_rows)
    # Slice input vector according to lag and saving into vessels
    for i_row in range(n_rows):
        dataX[i_row] = dataset[i_row:i_row+lag]
        dataY[i_row] = dataset[i_row+lag]
    # Return populated vessels
    return dataX, dataY
````

````python
nSplits = 5
nb = range(4,13,2)
lag = 3
kf = KFold(n_splits=nSplits)

rawData = dataframe.values

print("-"*63)

for n in nb:
  print("Numero de bloques {0}".format(n))
  print("-"*63)
  print("{0:^20} | {1:^20} | {2:^20}".format("Training Time [s]", "Training Score [ºC]", "Test Score [ºC]"))
  print("-"*63)

  trainingTime = 0
  trainScore = 0
  testScore = 0
  
  for train_index, test_index in kf.split(rawData):
    # Data preparation
    dfTrain, dfTest = rawData[train_index], rawData[test_index]
    dfTrain_scaled = scaler.transform(dfTrain)
    dfTest_scaled = scaler.transform(dfTest)
    trainX, trainY = create_dataset(dfTrain_scaled, lag)
    testX, testY = create_dataset(dfTest_scaled, lag)
    trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
    testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
    _trainY = scaler.inverse_transform([trainY])
    _testY = scaler.inverse_transform([testY])
    
    # Model training
    model = Sequential()
    model.add(LSTM(n, input_shape=(None, lag), activation='tanh', recurrent_activation='sigmoid'))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    tStart = time.clock()
    model.fit(trainX, trainY, epochs=25, batch_size=1, verbose=0)
    trainTime = time.clock() - tStart
    trainingTime+=trainTime
    
    # Prediction
    trainPredict = model.predict(trainX,batch_size=32)
    trainPredict = scaler.inverse_transform(trainPredict)
    testPredict = model.predict(testX,batch_size=32)
    testPredict = scaler.inverse_transform(testPredict)
    
    # Calculate RMSE
    trainS = np.sqrt(mean_squared_error(_trainY[0], trainPredict[:,0]))
    trainScore += trainS
    testS = np.sqrt(mean_squared_error(_testY[0], testPredict[:,0]))
    testScore += testS
    
    print("{0:^20.3f} | {1:^20.3f} | {2:^20.3f}".format(trainTime, trainS, testS))
    
  trainingTime/=nSplits  
  trainScore/=nSplits
  testScore/=nSplits
  print("-"*63)
  print("{0:^20.3f} | {1:^20.3f} | {2:^20.3f}".format(trainingTime, trainScore, testScore))
  print("-"*63)
  
  
  
````

Dicho de otra forma, para los $N_t$ datos/mediciones del conjunto de pruebas, se requiere predecir los $N_t-t$ valores a partir de algun valor $p$, de la siguiente manera:
$$
\hat{x}_{p} = f(x_{p-t},x_{p-t+1},\cdots,x_{p-1}) \\
\hat{x}_{p+1} = f(x_{p+1-t},x_{p+1-t+1},\cdots,x_{p-1},\hat{x}_{p}) \\
\hat{x}_{p+2} = f(x_{p+2-t},x_{p+2-t+1},\cdots,\hat{x}_{p},\hat{x}_{p+1}) \\
\cdots \\
\hat{x}_{N_t} = f(\hat{x}_{N_t-t},\hat{x}_{N_t-t+1},\cdots,\hat{x}_{N_t-2},\hat{x}_{N_t-1})
$$
